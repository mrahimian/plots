```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])
plt.plot(xpoints,ypoints)
```

```python
# from 0 to 20 with 0.1 sharpness 
xvals = np.arange(0,10,0.1)
plt.plot(xvals , np.sin(xvals))
# label axes
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
plt.plot(xvals , np.sin(xvals) , label = 'Sin X' , linestyle = '--')
plt.plot(xvals , np.cos(xvals) , label = 'Cos X' , color = 'green')
plt.legend()
```

```python
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma=10)
print(rands2d)
```

```python
plt.imshow(rands2d)
plt.colorbar()
```
